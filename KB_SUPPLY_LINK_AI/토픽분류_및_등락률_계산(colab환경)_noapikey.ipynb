### 사업설명 뽑아내는 함수

```python
import pandas as pd

def get_des_stock_name(df, stock_name):

    # 조건에 해당하는 행 필터링
    filtered_df = df[(df['stock_name'] == stock_name) &
                     (df['title'] == '1. 사업의 개요')]

    # 필터링된 행들의 'cleaned_text' 컬럼 내용 가져오기
    cleaned_texts = filtered_df['cleaned_text'].tolist()

    # 모든 텍스트를 합치기
    combined_text = ' '.join(cleaned_texts)

    return combined_text

```

### 사업설명 키워드 추출 (GPT)

```python
import openai

# OpenAI API 키 설정
openai.api_key = ''

def generate_topics(article_title):
    prompt = f"제목: {article_title}\n이 기업의 사업 내용을 가장 잘 표현할 수 있는 키워드 5가지를 뽑아줘 #로 키워드만 알려줘"

    response = openai.ChatCompletion.create(
      model="gpt-4",  # GPT-4 모델 사용
      messages=[
          {"role": "system", "content": "You are a helpful assistant."},
          {"role": "user", "content": prompt}
      ],
      max_tokens=2000,
      temperature=0.3
    )

    topics = response['choices'][0]['message']['content'].strip()
    return topics

def split_text(text, max_length=2000):
    """
    긴 텍스트를 일정 길이로 나누는 함수.
    """
    return [text[i:i + max_length] for i in range(0, len(text), max_length)]

# 특정 기업의 텍스트를 합친 후, 텍스트를 분할
txt_link = get_des_stock_name(all_df, '현대차')
text_segments = split_text(txt_link)

# 분할된 텍스트 각각에 대해 토픽 생성
all_results = []
for segment in text_segments:
    result = generate_topics(segment)
    all_results.append(result)

# 모든 결과를 합치기
final_result = " ".join(all_results)

# 결과 출력
print(final_result)
```

### 2.1  주 기업 크롤링 -> fnnews 크롤링

**주기업 : 현대차**

전처리 : 해당 기업명, 속보, 숫자 제외

```python
import re

def remove_company_name(titles, stok_name_):
    return [title.replace(stok_name_, '').strip() for title in titles]

def remove_company_name_and_extras(titles, stok_name_):
    cleaned_titles = []
    for title in titles:
        # Remove the stock name
        cleaned_title = title.replace(stok_name_, '').strip()
        # Remove the word '속보'
        cleaned_title = cleaned_title.replace('속보', '').strip()
        # Remove any numbers
        cleaned_title = re.sub(r'\d+', '', cleaned_title)
        # Remove extra spaces
        cleaned_title = ' '.join(cleaned_title.split())
        cleaned_titles.append(cleaned_title)
    return cleaned_titles
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 기본 URL 설정
base_url = 'https://www.fnnews.com/search'
params = {
    'page': '',
    'search_type': 'accuracy',
    'cont_type': 'tit',
    'period_type': 'direct',
    'searchDateS': '2024-04-01',
    'searchDateE': '2024-07-28',
    'search_txt': '현대차'
}

# 페이지 수 설정 (예시로 3페이지까지 크롤링)
page_limit = 10

# 크롤링 함수
def crawl_news(base_url, params, page_limit):
    news_data = []
    for page in range(1, page_limit + 1):
        params['page'] = page
        # 페이지 요청 및 파싱
        response = requests.get(base_url, params=params)
        soup = BeautifulSoup(response.content, 'html.parser')

        # 기사 제목과 날짜 찾기
        articles = soup.select('ul.list_article > li')

        for article in articles:
            title_tag = article.select_one('strong.tit_thumb a')
            date_tag = article.select_one('span.caption')

            if title_tag and date_tag:
                title = title_tag.text.strip()
                date = date_tag.text.strip()
                news_data.append((title, date))

    return news_data

# 뉴스 데이터 크롤링
news_data = crawl_news(base_url, params, page_limit)

# 데이터프레임 생성
df_news = pd.DataFrame(news_data, columns=['기사제목', '날짜'])

# 데이터프레임 출력
print(df_news)

# 데이터 저장 (선택 사항)
df_news.to_csv('samsung_securities_news.csv', index=False)
hyundae_news = df_news.copy()
```

### 2.4 경쟁사를 대상으로 토픽간 유사도 계산

```python
# SentenceTransformer 모델 로드
model = SentenceTransformer('distiluse-base-multilingual-cased-v2')

# 각 기사의 토픽을 임베딩 벡터로 변환
merged_df['임베딩'] = merged_df['키워드'].apply(lambda x: model.encode(x))

# 유사도 계산
similarity_matrix = cosine_similarity(merged_df['임베딩'].tolist())

# 유사한 기사들끼리 묶어서 결과 출력
results = []
for i in range(len(merged_df)):
    for j in range(i + 1, len(merged_df)):
        similarity = similarity_matrix[i][j]
        if similarity > 0.6:  # 유사도 기준은 0.5로 설정 (필요에 따라 조정 가능)
            results.append({
                '기사': merged_df['기사제목'][i],
                '기사 날짜': merged_df['날짜'][i],
                '기사 토픽': merged_df['키워드'][i],
                '유사 기사': merged_df['기사제목'][j],
                '유사 기사 날짜': merged_df['날짜'][j],
                '유사 기사 토픽': merged_df['키워드'][j],
                '유사도': similarity,
                '타겟기업의 주가 등락률%': merged_df['변동 %'][j]
            })

result_df = pd.DataFrame(results)
print(result_df)
```

### 2.5 경쟁사 유사 그룹으로 묶여진 토픽별 타겟기업의 주가 등락률 계산

```python
result_df2[['경쟁사 기사', '유사 기사','기사 토픽','유사도', '타겟기업의 주가 등락률%']]
```

```python
# 각 기사 A에 대해 유사 기사들의 변동성 평균 계산
average_volatility_results = []

# 각 기사 A별로 유사 기사들의 변동성을 계산
for article in result_df['기사'].unique():
    related_articles = result_df[result_df['기사'] == article]

    if not related_articles.empty:
        # 유사 기사들의 변동성 평균 계산
        avg_volatility = related_articles['타겟기업의 주가 등락률%'].mean()
    else:
        avg_volatility = None

    # 결과 저장
    average_volatility_results.append({
        '기사': article,
        '기사 날짜': related_articles['기사 날짜'].iloc[0],  # 같은 기사이므로 첫 번째 값을 사용
        '기사 토픽': related_articles['기사 토픽'].iloc[0],
        '타겟 기업의 평균 주가 등락률': avg_volatility
    })

# 최종 결과 데이터프레임
final_df = pd.DataFrame(average_volatility_results)
```
